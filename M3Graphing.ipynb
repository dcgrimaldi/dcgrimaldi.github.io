```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
df.head()
```

```python
fig,ax = plt.subplots()
ax.plot(df['TotalConstruction'],label='Total Construction',color='b')
ax.set_xlabel('Month')
ax.set_ylabel('Total Constructions')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid(True)
ax.set_facecolor("yellow")
```

```python
top_5 = df.nlargest(5, ['TotalConstruction'])
top_5.head()
```

```python
top_5.plot(
    x="MonthYear", y=["PrivateConstruction", "PublicConstruction"], kind="bar"
)
plt.title("Top Construction Dates",fontsize='14',fontname='Times New Roman')
plt.xlabel("Month&Year",fontsize='14',fontname='Times New Roman')
plt.ylabel("Construction Units",fontsize='14',fontname='Times New Roman')
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.axvspan(0,80000,ymin=0,ymax = 300,facecolor='k',alpha=0.1)
ax.text(100,85,'Summer of 06 & 07 saw the most construction',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
```

Looking at the total data the graph below shows how public construction increased as private construction dropped which then saw a decrease in public construction when private construction began to turn around - as shown in months 60 - 120.

```python
fig,ax = plt.subplots()
ax.plot(df['PrivateConstruction'],label='Private Construction')
ax.plot(df['PublicConstruction'],label='Public Construction')
ax.set_title("Total Constructions",fontsize=14,fontname='Times New Roman')
ax.set_xlabel('Month')
ax.set_ylabel('Construction Units')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.grid(True)
ax.set_facecolor("yellow")
ax.axvspan(60,120,ymin=0,ymax = 300,facecolor='k',alpha=0.2)
ax.text(80,100000,'Public Construction increased while Private Construction dropped',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
```
